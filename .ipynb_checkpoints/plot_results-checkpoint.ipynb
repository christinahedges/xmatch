```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pickle
import sys
import xmatch
```

```python
results=pd.read_pickle('results_probabilities.p')
```

```python
pos=results.Xflag=='gri'
plt.scatter(np.log10(results.EPICd2d)[pos],np.log10(results.PROB[pos]),s=1,label='gri')
pos=results.Xflag!='gri'
plt.scatter(np.log10(results.EPICd2d)[pos],np.log10(results.PROB[pos]),s=1,label='!gri')

plt.axvline(np.log10(6),c='black')
plt.axhline(-7,c='black')
plt.xlabel('Distance to Best Match')
plt.ylabel('Probability')
plt.legend()

threshold=-7
```

```python
#Force bad matches 
bad=np.arange(len(results))[np.any(([results.Nth_Neighbour==0],[np.log10(results.PROB)<=threshold]),axis=0)[0]]

results.PROB[bad]=0
results.loc[bad,'Nth_Neighbour']=0
results.loc[bad,'PROB']=0
results.loc[bad,'Xflag']='bad'
results.loc[bad,'EPICID']=0
results.loc[bad,'EPICd2d']=99
```

```python
_=plt.hist(results.InputMag[bad],80)
plt.xlabel('Magnitude')
plt.title('Bad Matches')
```

```python
print len(results),'targets'
print len(results[results.Nth_Neighbour>=1]),'targets matched well'
print len(results[results.Nth_Neighbour==0]),'targets with bad matching'
print len(results.Nth_Neighbour[results.Nth_Neighbour>1]),'where nearest neighbour is suboptimal'
```

```python
bad=np.where(results.Nth_Neighbour==0)[0]
pos=bad
print len(pos),'targets'
for j,i in enumerate(np.random.randint(0,len(pos),10)):
    s=xmatch.starmap(pos[i],results)
#    s.savefig('starmap{}.png'.format(j),dpi=100,bbox_inches='tight')

xmatch.starmap(0,results)
```
